```python
mu = 6
p = 0.8

def chicken_egg():
    n = Poisson(mu).draw()
    if (n > 0):
        x = Binomial(n, p).draw()
    else:
        x = 0
    return n, x

N, X = RV(ProbabilitySpace(chicken_egg))
Y = N - X

xy = (X & Y).sim(10000)
xy.plot('tile')
```

```python
x = xy[0]
x.plot()
Poisson(mu * p).plot()
```

```python
y = xy[1]
y.plot()
Poisson(mu * (1 - p)).plot()
```

```python
(X | (Y == 1) ).sim(1000).plot()
(X | (Y == 3)).sim(1000).plot('density')
```
