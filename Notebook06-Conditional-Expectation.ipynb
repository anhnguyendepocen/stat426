## Exer 6.1

```python
P = BoxModel([1, 2, 3, 4], size = 2)
X = RV(P, sum)
Y = RV(P, max)

xy = (X & Y).sim(10000)

xy.plot('tile')
```

```python
x = (X | (Y == 4) ).sim(10000)
x.plot()
x.mean(), x.var()
```

```python
x = (X | (Y == 3) ).sim(10000)
x.plot()
x.mean(), x.var()
```

```python
x = (X | (Y == 2) ).sim(10000)
x.plot()
x.mean(), x.var()
```

```python
x = (X | (Y == 1) ).sim(10000)
x.plot()
x.mean(), x.var()
```

```python
print('y', 'E(X|Y=y)')

for y in [1, 2, 3, 4]:
    x = (X | (Y == y) ).sim(10000).mean()
    print(y, x)
    
```

```python
x = X.sim(10000)
x.plot()
x.mean()
```

```python
print('y', 'E(X|Y=y)', 'E(XY|Y=y)')

for y in [1, 2, 3, 4]:
    x = (X | (Y == y) ).sim(10000).mean()
    z = (X * Y | (Y == y) ).sim(10000).mean()
    print(y, x, z)
```

```python
print('y', 'E(X|Y=y)', 'Var(X|Y=y)')

for y in [1, 2, 3, 4]:
    x = (X | (Y == y) ).sim(10000)
    print(y, x.mean(), x.var())
```

## Exer 7.2

```python
n = 10

def Uniform_Binomial():
    x = Uniform(0, 1).draw()
    y = Binomial(n, x).draw()
    return x, y

X, Y = RV(ProbabilitySpace(Uniform_Binomial))

(X & Y).sim(100).plot()
```

```python
xy = (X & Y).sim(10000)
xy.plot(jitter=True)
xy.cov(), xy.corr()
```

```python
y = xy[1]
y.plot()
y.mean(), y.var()
```

```python
x = xy[0]
(x * y).plot()
(x * y).mean()
```
