## Exer 4.2

```python
sigma = 10

def Normal_score(x, mu):
    return -(x - mu) / sigma ** 2

mu0 = 5

X = RV(Normal(mu0, sigma))
Y = X.apply(lambda x: Normal_score(x, mu0))
y = Y.sim(10000)
y.plot()
y.mean(), y.var()
```

```python
sigma = 0.1

def Normal_score(x, mu):
    return -(x - mu) / sigma ** 2

mu0 = 5

X = RV(Normal(mu0, sigma))
Y = X.apply(lambda x: Normal_score(x, mu0))
y = Y.sim(10000)
y.plot()
y.mean(), y.var()
```

## Exer 4.3

```python
def Poisson_score(x, mu):
    return -1 + x / mu

mu0 = 5

X = RV(Poisson(mu0))
Y = X.apply(lambda x: Poisson_score(x, mu0))
y = Y.sim(10000)
y.plot()
y.mean(), y.var()
```

```python
def Poisson_score2(x, mu):
    return  - x / mu ** 2

mu0 = 5

X = RV(Poisson(mu0))
Y = X.apply(lambda x: Poisson_score2(x, mu0))
y = Y.sim(10000)
y.plot()
-y.mean()
```

```python
n = 5
mu = 0
sigma = 3

P = Normal(mu, sigma) ** n
S2 = RV(P, var) * (n / (n - 1))
s = S2.sim(10000)

s.plot()
s.mean(), s.var(), 2 * sigma ** 4 / (n - 1), 2 * sigma ** 4 / n
```

```python
n = 5
mu = 0
sigma = 3

P = Normal(mu, sigma) ** n
Sigmahat2 = RV(P, var) 
s = Sigmahat2.sim(10000)

s.plot()
s.mean(), s.var(), 2 * sigma ** 4 / (n - 1) * ((n - 1) / n) ** 2
```

```python
theta = 2
n = 5

P = Uniform(0, theta) ** n
Thetahat = RV(P, max) * (n + 1) / n

t = Thetahat.sim(10000)
t.plot()
t.mean(), t.var(), theta ** 2 / (n * (n + 2)), theta ** 2 / n
```
