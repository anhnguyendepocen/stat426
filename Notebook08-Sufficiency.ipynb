```python
#@title Define Plotting Functions

def plot_discrete_function(f, xlim=(0, 10), xlabel=r"$\theta$", ylabel="Likelihood"):
    xs = np.arange(np.ceil(xlim[0]), np.floor(xlim[1]) + 1, dtype=int)
    ys = [f(x) for x in xs]
    plt.plot(xs, ys, "ko-")
    plt.xlabel(xlabel, fontsize=18)
    plt.ylabel(ylabel, fontsize=18)
    plt.xlim(*xlim)

def plot_continuous_function(f, xlim=(0, 1), xlabel=r"$\theta$", ylabel="Likelihood"):
    xs = np.linspace(xlim[0], xlim[1], 1000)
    ys = [f(x) for x in xs]
    plt.plot(xs, ys, "-")
    plt.xlabel(xlabel, fontsize=18)
    plt.ylabel(ylabel, fontsize=18)
    plt.xlim(*xlim)
```

## Exer 8.1

```python
n = 4

def Lik_a(p):
    return Binomial(n, p).pmf(1)

def Lik_b(p):
    return Binomial(2, p).pmf(1) * Binomial(2, p).pmf(0)

def Lik_c(p):
    return Bernoulli(p).pmf(0) * Bernoulli(p).pmf(1) * Bernoulli(p).pmf(0) * Bernoulli(p).pmf(0)

plot_continuous_function(Lik_a, xlim=(0, 1),
                         xlabel="$p$", ylabel="Likelihood")

plot_continuous_function(Lik_b, xlim=(0, 1),
                         xlabel="$p$", ylabel="Likelihood")

plot_continuous_function(Lik_c, xlim=(0, 1),
                         xlabel="$p$", ylabel="Likelihood")

plt.legend([r"$L_a(p)$", r"$L_b(p)$", r"$L_b(p)$"]);

plt.axvline(0.25, 0, 1, alpha=.3, linewidth=4);
```

```python
n = 4

def lik_a(p):
    return log(Binomial(n, p).pmf(1))

def lik_b(p):
    return log(Binomial(2, p).pmf(1) * Binomial(2, p).pmf(0))

def lik_c(p):
    return log(Bernoulli(p).pmf(0) * Bernoulli(p).pmf(1) * Bernoulli(p).pmf(0) * Bernoulli(p).pmf(0))

plot_continuous_function(lik_a, xlim=(0.001, 0.999),
                         xlabel="$p$", ylabel="Likelihood")

plot_continuous_function(lik_b, xlim=(0.001, 0.999),
                         xlabel="$p$", ylabel="Likelihood")

plot_continuous_function(lik_c, xlim=(0.001, 0.999),
                         xlabel="$p$", ylabel="Likelihood")

plt.legend([r"$\ell_a(p)$", r"$\ell_b(p)$", r"$\ell_b(p)$"]);

plt.axvline(0.25, 0, 1, alpha=.3, linewidth=4);
```
